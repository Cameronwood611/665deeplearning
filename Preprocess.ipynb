```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import cv2
import matplotlib.pyplot as plt
import os
```

```python
def size_image(folder):
    s = []
    for filename in os.listdir(folder):
        #print(f'{folder}/{filename}')
        try:
            im = cv2.imread(f'{folder}/{filename}')
            s.append(im.shape)
        except:
            pass
    return np.array(s)
```

```python
def max_size(shapes):
    a=max(shapes[:,0])
    b=max(shapes[:,1])
    return (a,b)
```

```python
folder1 = './sorted_images/general_equation'
equation = size_image(folder1)
print(max_size(equation))
```

```python
folder2 = './sorted_images/general_expression'
expression = size_image(folder2)
print(max_size(expression))
```

```python
folder3 = './sorted_images/inequality'
inequality = size_image(folder3)
print(max_size(inequality))
```

```python
folder4 = './sorted_images/not_equal'
noteq = size_image(folder4)
print(max_size(noteq))
```
